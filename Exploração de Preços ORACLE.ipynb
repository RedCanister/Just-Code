##Tratamento dos Dados

Apenas os 17 primeiros itens da lista foram divididos corretamente, o resto, est√° com erro de dados concentrado na coluna de 'Pre√ßos'. Tendo feito a divis√£o dessa coluna, temos um objeto onde na mesma linha, encontramos um Pandas Series, com o pre√ßo na posi√ß√£o [0], e o tipo de licen√ßa na posi√ß√£o [1] (ap√≥s a divis√£o da string com a fun√ß√£o do Python), os dois √∫nicos itens √∫teis para essa an√°lise.

```python
# A maior parte dos itens est√£o com duas colunas juntas em uma, na coluna de pre√ßo, que separamos nesse
#bloco, mas ainda precisamos remanejar os tipos de licen√ßa para sua pr√≥pria coluna, com o par√¢metro expand=True

precoSplit = precosDF['Pre√ßo'].str.split(" ", 1, expand=True)
precoSplit
```

### 6099 originais
-----
4464 restantes

O que equivale a 73,19% sendo usado na an√°lise comparada ao original, levando em considera√ß√£o os seguintes fatos: A tabela original estava no formato PDF.
```python
#Realizando o tratamento do dataset, excluindo linhas vazias e renomeando as colunas
precoClean = precoSplit.dropna()
precoClean = precoClean.reset_index()
precoClean = precoClean.drop(columns=['index'])
precoClean.columns = ['preco', 'tipo_licenca']
```

```python
# Convertendo (a hil√°ria) coluna de pre√ßos de string para float. A opera√ß√£o desse bloco de c√≥digo foi feito
#inplace, ou seja, sem gerar uma nova vari√°vel, ent√£o esse bloco poder√° ser executado uma √∫nica vez, ou
#com a cria√ß√£o de uma nova vari√°vel e DataFrame.

# Esse passo foi necess√°rio por que a fun√ß√£o .split() do Python sempre retorna 2 ou mais strings
precoClean['preco'] = [x.replace('.', '') for x in precoClean['preco']]
precoClean['preco'] = precoClean['preco'].apply(lambda x: float(x.split()[0].replace(',', '.')))
precoClean
```

```python
#Fazendo o mesmo para os tipos de licen√ßa, retirando os espa√ßos em branco das vari√°veis
precoClean['tipo_licenca'] = [x.replace('Manuten√ß√£o na Vig√™ncia', 'manutencao') for x in precoClean['tipo_licenca']]
precoClean['tipo_licenca'] = [x.replace('Permanente sem Manuten√ß√£o', 'permanente') for x in precoClean['tipo_licenca']]
precoClean['tipo_licenca'] = [x.replace('Suporte', 'suporte') for x in precoClean['tipo_licenca']]
precoClean['tipo_licenca'] = [x.replace('Assinatura', 'assinatura') for x in precoClean['tipo_licenca']]

# At√© eu descobrir que n√£o havia removido o espa√ßo em branco, apenas reduzido a label dos tipos de licen√ßa
# Mais sobre isso abaixo 
```

### Dados gerais do conjunto de dados
A partir daqui, faremos a explora√ß√£o do DataFrame com SQL, construindo tabelas de acordo com cada Tipo de Licen√ßa que podemos encontrar: 

1.   Permanente sem Manuten√ß√£o
2.   Manuten√ß√£o na Vig√™ncia
3.   Suporte
4.   Assinatura

```python
# describe() √© uma fun√ß√£o do pandas que pode rapidamente demonstrar todas as caracter√≠sticas
#estat√≠sticas do DataFrame selecionado. Observar que a contagem dessa fun√ß√£o, por ser float
#tamb√©m foi formatada como um pre√ßo
precoClean.describe()
```

```python
# √â com essa fun√ß√£o an√¥nima que vamos realizar os queries em SQL em cima do Dataframe, para n√£o ter que escrever a fun√ß√£o completa toda vez que for chamada 
pysqldf = lambda q: sqldf(q, globals())
```

```python
# Consultando quantos itens temos para cada tipo de licen√ßa
q = """
    SELECT tipo_licenca, COUNT(*) as Total
    FROM precoClean
    GROUP BY tipo_licenca
    HAVING COUNT(*)
    """

precoTotalItem = pysqldf(q)
precoTotalItem
```

```python
# Consultando qual o valor total para cada tipo de licen√ßa
q = """
    SELECT tipo_licenca, SUM(preco) as Total
    FROM precoClean
    GROUP BY tipo_licenca
    """

precoTotalValor = pysqldf(q)
precoTotalValor
```

```python
# Aqui observamos a m√©dia para cada tipo de licen√ßa
precoMediaA = precoTotalValor['Total'][0] / precoTotalItem['Total'][0]
precoMediaM = precoTotalValor['Total'][1] / precoTotalItem['Total'][1]
precoMediaP = precoTotalValor['Total'][2] / precoTotalItem['Total'][2]
precoMediaS = precoTotalValor['Total'][3] / precoTotalItem['Total'][3]

print('Pre√ßo m√©dio das licen√ßas permanentes: R$%.3f' % (precoMediaP))
print('Pre√ßo m√©dio das licen√ßas de manuten√ß√£o: R$%.3f' % (precoMediaM))
print('Pre√ßo m√©dio das licen√ßas de suporte: R$%.3f' % (precoMediaS))
print('Pre√ßo m√©dio das licen√ßas assinaturas: R$%.3f' % (precoMediaA))
```

Aqui realizamos a consulta para separarmos todas as licen√ßas desse tipo espec√≠fico, o que repetiremos para todos os outros tipos de licen√ßa. E para cada tipo de licen√ßa, selecionamos os 7 maiores e menores itens da lista de acordo com seu pre√ßo, e realizamos uma √∫ltima consulta no final, com amostras aleat√≥rias para cada tipo de licen√ßa, realizando assim a t√£o esperada explora√ß√£o dos pre√ßos das licen√ßas da Oracle.

### Tipo de licen√ßa Permanente sem Manuten√ß√£o

```python
q = """
    SELECT preco, tipo_licenca FROM precoClean
    WHERE tipo_licenca = 'permanente'
    """

precoPermanente = pysqldf(q)
precoPermanente.sort_values(by=['preco'], inplace=True)
precoPermanente.reset_index(inplace=True)
precoPermanente.drop(columns=['index'], inplace=True)
precoPermanente
```

```python
precoPH = precoPermanente['preco'].head(7)
precoPT = precoPermanente['preco'].tail(7)
dados = zip(precoPH, precoPT)

tabelaPermanente = pd.DataFrame(dados, columns=['bot7', 'top7'])
tabelaPermanente
```

```python
precoPermanente.sample(10, random_state=0)
```
